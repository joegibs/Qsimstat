```python
import sys
sys.path.append("../")

import numpy as np
from quimb import *
import quimb.tensor as qtn

from matplotlib import pyplot as plt


from scipy.stats import ks_2samp,rv_discrete
```

## Uniform bit strings

```python
N=10
m = 1000 #number of samples
n = 1000 #number of samples 2
def rand_bit_string(N,m):
    max_int = 2**N
    random_integers = np.random.randint(low=0, high=max_int, size=m)
    # Convert each integer to its binary string representation
    bitstrings = [i for i in random_integers]
    return bitstrings

sequence_uni: np.ndarray = rand_bit_string(N,m)
```

## PT random bit strings

```python
p=[]
for i in range(2**N):
    p.append((np.random.normal(0,0.2/(2**N),2)**2).sum())
p = np.array(p)
p = p/p.sum()
```

```python
custm = rv_discrete(name='custm', values=(np.arange(2**N), p))
sequence_PT = custm.rvs(size=m)
```

## haar random from mps sampling

```python

haar_mps = qtn.tensor_1d.MatrixProductState.from_dense(rand_haar_state(2**N))
sequence_haar_mps: np.ndarray = [int(''.join(np.array(i[0]).astype(str)),2) for i in haar_mps.sample(n)]
```

# Kolmogorovâ€“Smirnov

```python
ks_2samp(sequence_uni,sequence_haar_mps)
```

High P_value reject that they are the same dist

```python
ks_2samp(sequence_PT,sequence_haar_mps)
```

 low p_value (<0.1) accept that they are the same dist with 90% confidence
